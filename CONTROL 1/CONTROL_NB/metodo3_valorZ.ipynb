```r
install.packages("dplyr")
install.packages("ggplot2")
library("dplyr")
library("ggplot2")
```

```r
#carga archivo
datosPaises <- read.csv("https://raw.githubusercontent.com/fcarcamo03/PROG/main/CONTROL%201/DatosPaises.csv", sep = ";", header = TRUE)
 head(datosPaises)
```

```r
#INTERNET sin datos atipicos

datosPaises$Error <- abs(datosPaises$INTERNET-median(datosPaises$INTERNET))
datosPaises$ValorZ <- datosPaises$Error/median(datosPaises$Error)

  
datosPaises <- datosPaises %>%
            mutate(
                outlier = ifelse( ValorZ >= 4.5, "Outlier", "No outlier" )
            )
head(datosZ)
```

```r
datos_outliers <- datosPaises %>% filter( outlier == "No outlier")
```

```r
grafico_variabilidad <- datosPaises %>%
  ggplot() +
  aes(x = INTERNET, fill = outlier) +
  geom_histogram(binwidth = 1.5) +
  scale_fill_manual(values = c("No outlier" = "#3057D9", "Outlier" = "#4AF0BF"))+
  theme(panel.background = element_rect(fill = "white"),
        panel.grid.major = element_line(color = "gray", linetype = "dashed"))+
  ggtitle(label = "Metodo Valor-Z Robusto")+
  xlab("Internet") +
  ylab("N° de registros")+
  theme(plot.title = element_text(hjust = 0.5))

ggsave("internet_valorZ.svg.png", plot = grafico_variabilidad, width = 8, height = 6, dpi = 300)

grafico_variabilidad
```
